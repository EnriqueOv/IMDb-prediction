```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
nltk.download('punkt')
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
pd.set_option('display.max_colwidth',9999)
```

```python
movieset = pd.read_csv('moviesetwratings.csv')
ps = PorterStemmer()
```

```python
#porter stemmer

porterstem = movieset

for s in porterstem['summaries']:
  x = 0
  stemmed = word_tokenize(s)
  for y in stemmed:
    stemmed[x] = ps.stem(y)
    x = x + 1
  string = " ".join(stemmed)
  porterstem = porterstem.replace([s],string)
```

```python
#Bag of words for the porter stem of summaries
CountSumP = CountVectorizer(ngram_range=(1,1),stop_words='english',min_df=7)
CountSumDataP = CountSumP.fit_transform(porterstem['summaries'])
```

```python
porterBOW = pd.DataFrame(CountSumDataP.toarray(),columns=CountSumP.get_feature_names())
```

```python
#adds genres at the end
porterBOW['genre 1'] = movieset['genre1']
porterBOW['genre 2'] = movieset['genre2']
porterBOW['genre 3'] = movieset['genre3']
porterBOW['name'] = movieset['titles']
```

```python
#
porterBOW.to_csv('PorterBagOfWordsmin7.csv', index= False)
```
